```robotframework
*** Settings ***


Library           Autosphere.Browser
Library           Process
Library           Autosphere.Excel.Files
Library           BuiltIn
Test Teardown     Close All Browsers
Library           Collections
Library           Autosphere.Tables
Library           parse_table.py
Library           Autosphere.Archive
Library           Autosphere.FileSystem
Library    String
Resource    locator.robot


*** Keywords ***
Open The RPA Website
     Open Available Browser    https://botsdna.com/ActiveLoans/

Extract Link And Parse Data
    [Arguments]  ${file_link}  ${List}
    Log To Console  ${file_link}
            # Link Parsing Code 
            IF  '${file_link}' != '[]'
                ${link}=  Get Element Attribute  ${file_link}  href
                Run Process    curl  -o  ./Records/zip_file.zip  ${link}
                Extract Archive    ./Records/zip_file.zip  ./Records/
                ${Account_number}=  Set Variable  ${link.split('/')[-1].split('.')[0]}
                Log To Console    ${link}
                ${file}=  Read File    ./Records/${Account_number}.txt
                ${lines}=  Split To Lines  ${file}

                FOR    ${line}    IN    @{lines}
                    IF    '${line}' != ''
                         Log    ${line}
                         ${sp_text}=  Set Variable  ${line.split(':')[-1]}
                         ${label}=  Set Variable  ${line.split(':')[0]}
                         Append To List  ${List}  ${sp_text}

                    END


                END
            END
Get Header Details

    ${header}=  Create List   'STATUS'	'LOAN CODE' 	'PAN NUMBER'   'Account Number'  'Bank'  'Branch'  'Loan Taken On'  'Amount'  'EMI(month)'
    [Return]   ${header}
Extract Table Data
    ${TRows}=   Get WebElements    css:tr
    @{Data_List}=	Create List
    FOR    ${index}    ${tr}    IN ENUMERATE    @{TRows}
        IF    ${index} >= 1 and ${index} <= 3

            ${tDetails}=    Set Variable    ${tr.find_elements("tag name", 'td')}
            @{List} =	Create List
            FOR    ${td}    IN    @{tDetails}

                ${file_link}=   Set Variable    ${td.find_elements("tag name", 'a')}
                ${td_text}=  Get Element Attribute  ${td}  innerText
                Append To List	${List}  ${td_text}
                Extract Link And Parse Data    ${file_link}  ${List}
                Log To Console    ${List}
            END
            Append To List	${Data_List}  ${List}

        END
    END
    Create Table For Excel    ${Data_List}
Export Table to Excel
    [Arguments]  ${table}
    Create Workbook    active_loans.xlsx
    ${export}=       Export table    ${table}  with_index= True

    Append Rows To Worksheet    ${export}
    Save Workbook



Create Table For Excel
    [Arguments]  ${Data_List}
    ${header}=  Get Header Details
    ${table}=  Create Table  ${Data_List}  columns=${header}

     Export Table to Excel  ${table}

*** Tasks ***
RPA Challenge Round 1
    Open The RPA Website
    Extract Table Data


    [Teardown]    Close All Browsers
```
